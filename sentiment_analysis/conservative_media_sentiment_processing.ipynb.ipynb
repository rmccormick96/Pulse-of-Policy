# This function extracts website name from image url from the Turk Database 

```python
from urllib.parse import urlparse

string = "https://thehill.com/wp-content/uploads/sites/2/2018/04/trumpdonald_tariffs_040318getty.jpg?w=1280"

string.split("/")

string.split("/")[2]

def extract_website_name(images_list):

    if images_list and len(images_list) > 0:
        url = images_list[0]['url']
    
        parsed_url = urlparse(url)
        domain_name = parsed_url.netloc
    else:
        return None
   
    return domain_name.replace('www.', '')

```

# These following function is to extract the data of the article from the image urls 

```python
from datetime import datetime
import re

def extract_date_from_url(images_list):
    if images_list and len(images_list) > 0:
        url = images_list[0]['url']
        
        date_patterns = [
            r'(\d{4})[/-](\d{2})[/-](\d{2})',  
            r'(\d{2})[/-](\d{2})[/-](\d{4})', 
            r'(\d{2})(\d{2})(\d{4})',         
            r'(\d{4})(\d{2})(\d{2})',         
            r'(\d{4})[/-](\d{2})'           
        ]
    
        for pattern in date_patterns:
            match = re.search(pattern, url)
            if match:
                groups = match.groups()
                try:
                    if len(groups) == 3:
                        
                        if len(groups[0]) == 4:  
                            year, month, day = groups
                        elif len(groups[2]) == 4: 
                            
                            month, day, year = groups
                            
                        elif len(groups[0]) == 2: 
                            if pattern == r'(\d{2})(\d{2})(\d{4})':  
                                month, day, year = groups
                            else: 
                                year, month, day = groups

                       
                        date = datetime.strptime(f"{year}-{month}-{day}", "%Y-%m-%d").date()
                        return date.strftime("%Y-%m-%d")
                except ValueError:
                    continue
        return "No logical date found"
    return None

```

```python
import re
from datetime import datetime

def extract_date_from_url_2(images_list):
    if images_list and len(images_list) > 0:

        for i in images_list:
            url =i["url"]
            
            # Adding a pattern to match YYYY/MM and assuming the first day of the month
            date_patterns = [
                r'(\d{4})[/-](\d{2})[/-](\d{2})',  # YYYY/MM/DD or YYYY-MM-DD
                r'(\d{2})[/-](\d{2})[/-](\d{4})',  # MM-DD-YYYY or DD-MM-YYYY
                r'(\d{4})[/-](\d{2})',             # YYYY/MM, assuming the first day of the month
            ]
        
            for pattern in date_patterns:
                match = re.search(pattern, url)
                if match:
                    groups = match.groups()
                    try:
                        year, month = groups[0], groups[1]
                        day = "01" if len(groups) < 3 else groups[2]  # Default to the first day if DD is missing

                        # Validate the date
                        date = datetime.strptime(f"{year}-{month}-{day}", "%Y-%m-%d").date()
                        return date.strftime("%Y-%m-%d")
                    except ValueError:
                        continue  # If the date is invalid, continue to the next pattern
        return "No logical date found"
    return None
```

# Sentiment Analysis 

```python
from transformers import pipeline, AutoTokenizer
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
revision = "af0f99b"
classifier = pipeline("sentiment-analysis", model=model_name, revision=revision)
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python

import numpy as np
import logging
from transformers import logging as transformers_logging
transformers_logging.set_verbosity_error()

def sentiment(text):
    model_name = "distilbert-base-uncased-finetuned-sst-2-english"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    classifier = pipeline("sentiment-analysis", model=model_name)

   
    max_length = 512 - 10  
    stride = 256

    tokens = tokenizer(text, add_special_tokens=False, return_tensors="pt", truncation=False)["input_ids"].squeeze()
    
    if len(tokens) <= max_length:
        results = [classifier(text)]  
    else:
        chunks = [tokens[i : i + max_length] for i in range(0, len(tokens), max_length - stride)]
        results = []


            
        for chunk in chunks:
            
            chunk_text = tokenizer.decode(chunk, skip_special_tokens=True)
          
            
            strip = chunk_text.split(" ")
            
            
            result = classifier(chunk_text)
            
            results.append(result)
    
    positive_scores = sum([r[0]['score'] for r in results if r[0]['label'] == 'POSITIVE'])
    
    negative_scores = sum([r[0]['score'] for r in results if r[0]['label'] == 'NEGATIVE'])
    return 1 if positive_scores >= negative_scores else 0


```
